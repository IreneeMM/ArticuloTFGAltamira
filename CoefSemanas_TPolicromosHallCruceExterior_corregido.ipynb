### Obtener coeficientes:

```python
#Función adaptada para variables sin suavizar
def cc_Th_Tc_Te_Tp(DataframeCruce, DataframeExterior, DataframeHall, DatosPolicromos):
    """
    Atributos:
    1) DataframeCruce: Lista de dataframes por semanas correspondiente a los datos de un mes de Cruce
    2) DataframeExterior: Lista de dataframes por semanas correspondiente a los datos de un mes de Exterior
    3) DataframeHall: Lista de dataframes por semanas correspondiente a los datos de un mes de Hall
    4) DatosPolicromos: Lista de dataframes por semanas correspondiente a los datos de un mes de Policromos
    
    Devuelve:
    Una lista de listas, r. Cada elemento de r, r[i] con i=0, 1, 2, 3, 4, es una lista con los coeficiente de correlación y p 
    valores correspondientes a una semana.  r[i] = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te
    
    ¿A qué corresponde cada elemento de r[i]?
    1) r_Tp_Th: Coeficiente de correlación entre la temperatura de Policromos y del Hall
    2) p_Tp_Th: P-valor correspondiente a r_Tp_Th
    3) r_Tp_Tc: Coeficiente de correlación entre la temperatura de Policromos y del Cruce
    4) p_Tp_Tc: P-valor correspondiente a r_Tp_Tc
    5) r_Tp_Te: Coeficiente de correlación entre la temperatura de Policromos y la exterior
    6) p_Tp_Te: P-valor correspondiente a r_Tp_Te
    """
    i=0
    r=[]
    while i<len(DataframeCruce):
       #print(i)
        DataframeCruce[i].reset_index(drop=True, inplace=True)
        DataframeExterior[i].reset_index(drop=True, inplace=True)
        DataframeHall[i].reset_index(drop=True, inplace=True)
        DatosPolicromos[i].reset_index(drop=True, inplace=True)
        
        # Variables que me van a hacer falta: Tp_Th, Tp_Tc, Tp_Te
        for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
            if  __builtins__.sum([pd.isna(DataframeCruce[i]["Temperatura D (° C)"][index]),
                                  pd.isna(DataframeHall[i]['Temperatura A (° C)'][index]),
                                  pd.isna(DataframeExterior[i]["Tr Exterior (C Tr)"][index]),
                                  pd.isna(DatosPolicromos[i]['Tp'][index])]) > 0:
                DataframeCruce[i].drop([index], axis=0, inplace=True)
                DataframeExterior[i].drop([index], axis=0, inplace=True)
                DataframeHall[i].drop([index], axis=0, inplace=True)
                DatosPolicromos[i].drop([index], axis=0, inplace=True)

        #Mediante scipy por semanas:
        r_Tp_Th, p_Tp_Th = stats.pearsonr(DatosPolicromos[i]['Tp'], DataframeHall[i]['Temperatura A (° C)'])
        r_Tp_Tc, p_Tp_Tc = stats.pearsonr(DatosPolicromos[i]['Tp'], DataframeCruce[i]["Temperatura D (° C)"])
        r_Tp_Te, p_Tp_Te = stats.pearsonr(DatosPolicromos[i]['Tp'], DataframeExterior[i]["Tr Exterior (C Tr)"])
        rs = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te, len(DatosPolicromos[i]['Tp'])
        r.append(rs)
        i=i+1
    return(r)
```

```python
#Función adaptada para variables suavizadas
def cc_Th_Tc_Te_Tp(DataframeCruce, DataframeExterior, DataframeHall, DatosPolicromos):
    """
    Atributos:
    1) DataframeCruce: Lista de dataframes por semanas correspondiente a los datos de un mes de Cruce
    2) DataframeExterior: Lista de dataframes por semanas correspondiente a los datos de un mes de Exterior
    3) DataframeHall: Lista de dataframes por semanas correspondiente a los datos de un mes de Hall
    4) DatosPolicromos: Lista de dataframes por semanas correspondiente a los datos de un mes de Policromos
    
    Devuelve:
    Una lista de listas, r. Cada elemento de r, r[i] con i=0, 1, 2, 3, 4, es una lista con los coeficiente de correlación y p 
    valores correspondientes a una semana.  r[i] = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te
    
    ¿A qué corresponde cada elemento de r[i]?
    1) r_Tp_Th: Coeficiente de correlación entre la temperatura de Policromos y del Hall
    2) p_Tp_Th: P-valor correspondiente a r_Tp_Th
    3) r_Tp_Tc: Coeficiente de correlación entre la temperatura de Policromos y del Cruce
    4) p_Tp_Tc: P-valor correspondiente a r_Tp_Tc
    5) r_Tp_Te: Coeficiente de correlación entre la temperatura de Policromos y la exterior
    6) p_Tp_Te: P-valor correspondiente a r_Tp_Te
    """
    i=0
    r=[]
    while i<len(DataframeCruce):
       #print(i)
        DataframeCruce[i].reset_index(drop=True, inplace=True)
        DataframeExterior[i].reset_index(drop=True, inplace=True)
        DataframeHall[i].reset_index(drop=True, inplace=True)
        DatosPolicromos[i].reset_index(drop=True, inplace=True)
        
        # Variables que me van a hacer falta: Tp_Th, Tp_Tc, Tp_Te
        for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
            if  __builtins__.sum([pd.isna(DataframeCruce[i]['Tc_suavizada'][index]),
                                  pd.isna(DataframeHall[i]['Th_suavizada'][index]),
                                  pd.isna(DataframeExterior[i]['Te_suavizada'][index]),
                                  pd.isna(DatosPolicromos[i]['Tp_suavizada'][index])]) > 0:
                DataframeCruce[i].drop([index], axis=0, inplace=True)
                DataframeExterior[i].drop([index], axis=0, inplace=True)
                DataframeHall[i].drop([index], axis=0, inplace=True)
                DatosPolicromos[i].drop([index], axis=0, inplace=True)

        #Mediante scipy por semanas:
        r_Tp_Th, p_Tp_Th = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeHall[i]['Th_suavizada'])
        r_Tp_Tc, p_Tp_Tc = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeCruce[i]['Tc_suavizada'])
        r_Tp_Te, p_Tp_Te = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeExterior[i]['Te_suavizada'])
        rs = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te, len(DatosPolicromos[i]['Tp_suavizada'])
        r.append(rs)
        i=i+1
    return(r)
```
