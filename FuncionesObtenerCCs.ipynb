```python
#Correlaciones de los datos sin suavizar
def cc_Tr_Tc_Te(DataframeCruce, DataframeExterior):#Función adaptada para variables sin suavizar
    """
    Atributos:
    1) DataframeCruce: Lista de dataframes por semanas correspondiente a los datos de un mes de Cruce
    2) DataframeExterior: Lista de dataframes por semanas correspondiente a los datos de un mes de Exterior
    
    Devuelve:
    Una lista de listas, r. Cada elemento de r, r[i] con i=0, 1, 2, 3, 4, es una lista con los coeficiente de correlación y p 
    valores correspondientes a una semana.  r[i] = r_TD_Tr, p_TD_Tr, r_TD_Te, p_TD_Te, r_Tr_Te, p_Tr_Te
    
    ¿A qué corresponde cada elemento de r[i]?
    1) r_TD_Tr: Coeficiente de correlación entre la temperatura del Cruce y del Ramal
    2) p_TD_Tr: P-valor correspondiente a r_TD_Tr
    3) r_TD_Te: Coeficiente de correlación entre la temperatura del Cruce y la exterior
    4) p_TD_Te: P-valor correspondiente a r_TD_Te
    5) r_Tr_Te: Coeficiente de correlación entre la temperatura del Ramal y la exterior
    6) p_Tr_Te: P-valor correspondiente a r_Tr_Te
    """
    i=0
    r=[]
    while i<len(DataframeCruce):
       #print(i)
        DataframeCruce[i].reset_index(drop=True, inplace=True)
        DataframeExterior[i].reset_index(drop=True, inplace=True)
        
        # Variables que me van a hacer falta: Tc, Te, Tr
        if DataframeCruce[i]["Temperatura D (° C)"].isna().all():
#             print('No datos en cruce')
            rs = np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, len(DataframeExterior[i]["Tr Exterior (C Tr)"])
        
        else:
#             print('si datos en cruce')
            for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
                if  __builtins__.sum([pd.isna(DataframeCruce[i]["Temperatura D (° C)"][index]),
                                      pd.isna(DataframeCruce[i]['Tr_suavizada'][index]),
                                      pd.isna(DataframeExterior[i]["Tr Exterior (C Tr)"][index])]) > 0:
                    DataframeCruce[i].drop([index], axis=0, inplace=True)
                    DataframeExterior[i].drop([index], axis=0, inplace=True)

            #Mediante scipy por semanas:
            r_TD_Tr, p_TD_Tr = stats.pearsonr(DataframeCruce[i]["Temperatura D (° C)"], DataframeCruce[i]["tr Cruce Ramal ('C)"])
            r_TD_Te, p_TD_Te = stats.pearsonr(DataframeCruce[i]["Temperatura D (° C)"], DataframeExterior[i]["Tr Exterior (C Tr)"])
            r_Tr_Te, p_Tr_Te = stats.pearsonr(DataframeCruce[i]["tr Cruce Ramal ('C)"], DataframeExterior[i]["Tr Exterior (C Tr)"])
            rs = r_TD_Tr, p_TD_Tr, r_TD_Te, p_TD_Te, r_Tr_Te, p_Tr_Te, len(DataframeCruce[i]["tr Cruce Ramal ('C)"])
        r.append(rs)
        i=i+1
    return(r)
```

```python
#Función adaptada a datos sin suavizar
def difT_CO2_cc_por_semanas(DataframeCruce, DataframeExterior, DataframeHall, PolicromosSemanas):
    """
    Atributos:
    1) DataframeCruce: Lista de dataframes por semanas correspondiente a los datos de un mes de Cruce
    2) DataframeExterior: Lista de dataframes por semanas correspondiente a los datos de un mes de Exterior
    3) DataframeHall: Lista de dataframes por semanas correspondiente a los datos de un mes de Hall
    4) DatosPolicromos: Lista de dataframes por semanas correspondiente a los datos de un mes de Policromos
    
    Devuelve:
    Una lista de listas, r. Cada elemento de r, r[i] con i=0, 1, 2, 3, 4, es una lista con los coeficiente de correlación y p 
    valores correspondientes a una semana.  r[i] = r_difT_CO2_cruce, p_difT_CO2_cruce, r_difT_CO2_hall, p_difT_CO2_hall, 
                                                   r_difT_CO2_policromos, p_difT_CO2_policromos
    
    ¿A qué corresponde cada elemento de r[i]?
    1) r_difT_CO2_cruce: Coeficiente de correlación entre la diferencia de temperaturas de Cruce y el exterior y el CO2 del 
    Cruce
    2) p_difT_CO2_cruce: P-valor correspondiente a r_difT_CO2_cruce
    3) r_difT_CO2_hall: Coeficiente de correlación entre la diferencia de temperaturas de Cruce y el exterior y el CO2 del Hall
    4) p_difT_CO2_hall: P-valor correspondiente a r_difT_CO2_hall
    5) r_difT_CO2_policromos: Coeficiente de correlación entre la diferencia de temperaturas de Cruce y el exterior y el CO2 de 
    Policromos
    6) p_difT_CO2_policromos: P-valor correspondiente a r_Tp_Te
    """
    i=0
    r=[]
    while i<len(DataframeCruce):
       #print(i)
        DataframeCruce[i].reset_index(drop=True, inplace=True)
        DataframeExterior[i].reset_index(drop=True, inplace=True)
        DataframeHall[i].reset_index(drop=True, inplace=True)
        PolicromosSemanas[i].reset_index(drop=True, inplace=True)
#         print(DataframeCruce[i]['Fecha/Tiempo'][0], i)
        # Variables que me van a hacer falta: Tc, Te, Th, Tp, CO2c, CO2h, CO2p
        if DataframeCruce[i]["Temperatura D (° C)"].isna().all() or DataframeCruce[i]['CO2 D (ppm)'].isna().all() :
#             print('No datos en cruce')
            for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
                if  __builtins__.sum([pd.isna(DataframeHall[i]['Temperatura A (° C)'][index]),
                                  pd.isna(DataframeHall[i]['CO2 A (ppm)'][index]),
                                  pd.isna(PolicromosSemanas[i]['Tp'][index]),
                                  pd.isna(PolicromosSemanas[i]["CO2"][index]),
                                  pd.isna(DataframeExterior[i]["Tr Exterior (C Tr)"][index])]) > 0:
                    DataframeCruce[i].drop([index], axis=0, inplace=True)
                    DataframeExterior[i].drop([index], axis=0, inplace=True) 
                    DataframeHall[i].drop([index], axis=0, inplace=True) 
                    PolicromosSemanas[i].drop([index], axis=0, inplace=True) 
            Dif_T_Hall = DataframeExterior[i]["Tr Exterior (C Tr)"]-DataframeHall[i]['Temperatura A (° C)']
            Dif_T_Policromos = DataframeExterior[i]["Tr Exterior (C Tr)"]-PolicromosSemanas[i]['Tp']
            r_difT_CO2_hall, p_difT_CO2_hall = stats.pearsonr(Dif_T_Hall, DataframeHall[i]["CO2 A (ppm)"])
            r_difT_CO2_policromos, p_difT_CO2_policromos = stats.pearsonr(Dif_T_Policromos, PolicromosSemanas[i]["CO2"])
            rs = np.nan, np.nan, r_difT_CO2_hall, p_difT_CO2_hall, r_difT_CO2_policromos, p_difT_CO2_policromos, len(DataframeExterior[i]["Tr Exterior (C Tr)"])
        
        else:
#             print('si datos en cruce')
            for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
                if  __builtins__.sum([pd.isna(DataframeCruce[i]["Temperatura D (° C)"][index]),
                                  pd.isna(DataframeCruce[i]['CO2 D (ppm)'][index]),
                                  pd.isna(DataframeHall[i]['Temperatura A (° C)'][index]),
                                  pd.isna(DataframeHall[i]['CO2 A (ppm)'][index]),
                                  pd.isna(PolicromosSemanas[i]['Tp'][index]),
                                  pd.isna(PolicromosSemanas[i]["CO2"][index]),
                                  pd.isna(DataframeExterior[i]["Tr Exterior (C Tr)"][index])]) > 0:
                    DataframeCruce[i].drop([index], axis=0, inplace=True)
                    DataframeExterior[i].drop([index], axis=0, inplace=True) 
                    DataframeHall[i].drop([index], axis=0, inplace=True) 
                    PolicromosSemanas[i].drop([index], axis=0, inplace=True) 
            #Quiero los coef entre CO2 y dif de temperaturas:
            Dif_T_cruce= DataframeExterior[i]["Tr Exterior (C Tr)"]-DataframeCruce[i]["Temperatura D (° C)"]
            Dif_T_Hall = DataframeExterior[i]["Tr Exterior (C Tr)"]-DataframeHall[i]['Temperatura A (° C)']
            Dif_T_Policromos = DataframeExterior[i]["Tr Exterior (C Tr)"]-PolicromosSemanas[i]['Tp']
            #Mediante scipy:
            r_difT_CO2_cruce, p_difT_CO2_cruce = stats.pearsonr(Dif_T_cruce, DataframeCruce[i]["CO2 D (ppm)"])
            r_difT_CO2_hall, p_difT_CO2_hall = stats.pearsonr(Dif_T_Hall, DataframeHall[i]["CO2 A (ppm)"])
            r_difT_CO2_policromos, p_difT_CO2_policromos = stats.pearsonr(Dif_T_Policromos, PolicromosSemanas[i]["CO2"])
            rs = r_difT_CO2_cruce, p_difT_CO2_cruce, r_difT_CO2_hall, p_difT_CO2_hall, r_difT_CO2_policromos, p_difT_CO2_policromos, len(DataframeExterior[i]["Tr Exterior (C Tr)"])
        r.append(rs)
        i=i+1
    return(r)
```

```python
#Función adaptada para variables suavizadas
def cc_Th_Tc_Te_TpSuav(DataframeCruce, DataframeExterior, DataframeHall, DatosPolicromos):
    """
    Atributos:
    1) DataframeCruce: Lista de dataframes por semanas correspondiente a los datos de un mes de Cruce
    2) DataframeExterior: Lista de dataframes por semanas correspondiente a los datos de un mes de Exterior
    3) DataframeHall: Lista de dataframes por semanas correspondiente a los datos de un mes de Hall
    4) DatosPolicromos: Lista de dataframes por semanas correspondiente a los datos de un mes de Policromos
    
    Devuelve:
    Una lista de listas, r. Cada elemento de r, r[i] con i=0, 1, 2, 3, 4, es una lista con los coeficiente de correlación y p 
    valores correspondientes a una semana.  r[i] = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te
    
    ¿A qué corresponde cada elemento de r[i]?
    1) r_Tp_Th: Coeficiente de correlación entre la temperatura de Policromos y del Hall
    2) p_Tp_Th: P-valor correspondiente a r_Tp_Th
    3) r_Tp_Tc: Coeficiente de correlación entre la temperatura de Policromos y del Cruce
    4) p_Tp_Tc: P-valor correspondiente a r_Tp_Tc
    5) r_Tp_Te: Coeficiente de correlación entre la temperatura de Policromos y la exterior
    6) p_Tp_Te: P-valor correspondiente a r_Tp_Te
    """
    i=0
    r=[]
    while i<len(DataframeCruce):
       #print(i)
        DataframeCruce[i].reset_index(drop=True, inplace=True)
        DataframeExterior[i].reset_index(drop=True, inplace=True)
        DataframeHall[i].reset_index(drop=True, inplace=True)
        DatosPolicromos[i].reset_index(drop=True, inplace=True)
        
        # Variables que me van a hacer falta: Tp_Th, Tp_Tc, Tp_Te
        if DataframeCruce[i]['CO2 D (ppm)'].isna().all() :
#             print('No datos en cruce')
            for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
                if  __builtins__.sum([pd.isna(DataframeHall[i]['Th_suavizada'][index]),
                                      pd.isna(DataframeExterior[i]['Te_suavizada'][index]),
                                      pd.isna(DatosPolicromos[i]['Tp_suavizada'][index])]) > 0:
                    DataframeExterior[i].drop([index], axis=0, inplace=True)
                    DataframeHall[i].drop([index], axis=0, inplace=True)
                    DatosPolicromos[i].drop([index], axis=0, inplace=True)

            #Mediante scipy por semanas:
            r_Tp_Th, p_Tp_Th = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeHall[i]['Th_suavizada'])
            r_Tp_Te, p_Tp_Te = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeExterior[i]['Te_suavizada'])
            rs = r_Tp_Th, p_Tp_Th, np.nan, np.nan, r_Tp_Te, p_Tp_Te, len(DatosPolicromos[i]['Tp_suavizada'])
        
        else:
#             print('si datos en cruce')
            for index, _ in DataframeExterior[i].iterrows():#Se va fuera de indice
                if  __builtins__.sum([pd.isna(DataframeCruce[i]['Tc_suavizada'][index]),
                                      pd.isna(DataframeHall[i]['Th_suavizada'][index]),
                                      pd.isna(DataframeExterior[i]['Te_suavizada'][index]),
                                      pd.isna(DatosPolicromos[i]['Tp_suavizada'][index])]) > 0:
                    DataframeCruce[i].drop([index], axis=0, inplace=True)
                    DataframeExterior[i].drop([index], axis=0, inplace=True)
                    DataframeHall[i].drop([index], axis=0, inplace=True)
                    DatosPolicromos[i].drop([index], axis=0, inplace=True)

            #Mediante scipy por semanas:
            r_Tp_Th, p_Tp_Th = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeHall[i]['Th_suavizada'])
            r_Tp_Tc, p_Tp_Tc = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeCruce[i]['Tc_suavizada'])
            r_Tp_Te, p_Tp_Te = stats.pearsonr(DatosPolicromos[i]['Tp_suavizada'], DataframeExterior[i]['Te_suavizada'])
            rs = r_Tp_Th, p_Tp_Th, r_Tp_Tc, p_Tp_Tc, r_Tp_Te, p_Tp_Te, len(DatosPolicromos[i]['Tp_suavizada'])
        r.append(rs)
        i=i+1
    return(r)
```
