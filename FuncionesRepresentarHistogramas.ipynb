# Cruce

```python
#Temperatura del cruce sin suavizar
def HistogramaTCruce(FicheroCruce, m, divisiones, MesAño, minVariable, maxVariable):
    '''
    Con atributos:
    1) Fichero Cruce: Fichero de datos de la sala
    2) m: La altura en el eje de ordenadas que se quiere representar
    3) divisiones: Número de intervalos en los que se quiere dividir el rango de datos de la variable
    4) MesAño: String indicando el mes y el año de los datos a representar
    5) minVariable: Minimo valor en el eje de abscisas a representar
    6) maxVariable: Minimo valor en el eje de abscisas a representar
    Se representa el histograma de la temperatura del Cruce por meses acotado según m, minVarible y 
    maxVariable.
    '''
    #Sacar el histograma
    df = pd.read_excel(FicheroCruce)
    bins = np.linspace(df["Temperatura D (° C)"].min(), df["Temperatura D (° C)"].max(), divisiones)
    np.histogram(df["Temperatura D (° C)"], bins=bins)
    plt.ylim(0, m)
    plt.xlim(minVariable,maxVariable)
    #Pintarlo+leyenda
    plt.grid()
    plt.hist(df["Temperatura D (° C)"], bins=divisiones, edgecolor='black')
    plt.title(MesAño)
    plt.xlabel("Temperatura Cruce / ºC")
    plt.ylabel("Frecuencia")
    plt.gca().set_axisbelow(True)
    plt.show()
```
