```python
m=0 #Máximo de datos por columna hasta este momento
def AlturaTCruce(FicheroCruce, m, divisiones):
#     print('---------------------------------------------------------------')
    df = pd.read_excel(FicheroCruce)
    bins = np.linspace(df["Temperatura D (° C)"].min(), df["Temperatura D (° C)"].max(), divisiones)
    hist, bins = np.histogram(df["Temperatura D (° C)"], bins=bins)
#     print('Máximo de datos por columna hasta ahora:', m)
    for i in range(len(hist)):
#         print(f"Intervalo {i+1}: {hist[i]} datos")
        if hist[i]>m:
            m=hist[i]
    return m
```

```python
def HistogramaTRamalsuavCruce(FicheroCruce, m, divisiones, MesAño):
    #Sacar el histograma
    df = pd.read_excel(FicheroCruce)
    bins = np.linspace(df['Tr_suavizada'].min(), df['Tr_suavizada'].max(), divisiones)
    np.histogram(df['Tr_suavizada'], bins=bins)
    plt.ylim(0, m)
    #Pintarlo+leyenda
    plt.grid()
    plt.hist(df['Tr_suavizada'], bins=divisiones, edgecolor='black')
    plt.title(MesAño)
    plt.xlabel("Temperatura suavizada Ramal / ºC")
    plt.ylabel("Frecuencia")
    plt.gca().set_axisbelow(True)
    plt.show()
```
