### Sacar el máximo de veces que se repite un dato en los histogramas:

```python
m=0 #Máximo de datos por columna hasta este momento
def AlturaTEntrada(FicheroEntrada, m, divisiones):
#     print('---------------------------------------------------------------')
    df = pd.read_excel(FicheroEntrada)
    bins = np.linspace(df['Temperatura A (° C)'].min(), df['Temperatura A (° C)'].max(), divisiones)
    hist, bins = np.histogram(df['Temperatura A (° C)'], bins=bins)
#     print('Máximo de datos por columna hasta ahora:', m)
    for i in range(len(hist)):
#         print(f"Intervalo {i+1}: {hist[i]} datos")
        if hist[i]>m:
            m=hist[i]
    return m
```
