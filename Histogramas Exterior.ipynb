# 1. Histogramas de la temperatura de Exterior sin suavizar


### Sacar el máximo de veces que se repite un dato en los histogramas:

```python
m=0 #Máximo de datos por columna hasta este momento
minVariable=100000000000#Minimo  para comparar de la variable a representar
maxVariable=0#Máximo para comparar de la variable a representar
def AlturaTExterior(FicheroExterior, m, divisiones, minVariable, maxVariable):
#     print('---------------------------------------------------------------')
    df = pd.read_excel(FicheroExterior)
    min_relativo=df["Tr Exterior (C Tr)"].min()
    max_relativo=df["Tr Exterior (C Tr)"].max()
    bins = np.linspace(min_relativo, max_relativo, divisiones)
    hist, bins = np.histogram(df["Tr Exterior (C Tr)"], bins=bins)
#     print('Máximo de datos por columna hasta ahora:', m)
    for i in range(len(hist)):
#         print(f"Intervalo {i+1}: {hist[i]} datos")
        if hist[i]>m:
            m=hist[i]
    #Igualar la escala en x:
    if min_relativo<minVariable:
        minVariable=min_relativo
    if max_relativo>maxVariable:
        maxVariable=max_relativo
    return m, minVariable, maxVariable
```

### Representar histogramas

```python
def HistogramaTExterior(FicheroExterior, m, divisiones, MesAño, minVariable, maxVariable):
    #Sacar el histograma
    df = pd.read_excel(FicheroExterior)
    bins = np.linspace(df["Tr Exterior (C Tr)"].min(), df["Tr Exterior (C Tr)"].max(), divisiones)
    np.histogram(df["Tr Exterior (C Tr)"], bins=bins)
    plt.ylim(0, m)
    plt.xlim(minVariable,maxVariable)
    #Pintarlo+leyenda
    plt.grid()
    plt.hist(df["Tr Exterior (C Tr)"], bins=divisiones, edgecolor='black')
    plt.title(MesAño)
    plt.xlabel("Temperatura Exterior / ºC")
    plt.ylabel("Frecuencia")
    plt.gca().set_axisbelow(True)
    plt.show()
```
