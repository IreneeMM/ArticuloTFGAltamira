```python
m=0 #Máximo de datos por columna hasta este momento
def AlturaCO2suavPolicromos(FicheroPolicromos, m, divisiones):
#     print('---------------------------------------------------------------')
    df = pd.read_excel(FicheroPolicromos)
    bins = np.linspace(df['CO2p_suavizado'].min(), df['CO2p_suavizado'].max(), divisiones)
    hist, bins = np.histogram(df['CO2p_suavizado'], bins=bins)
#     print('Máximo de datos por columna hasta ahora:', m)
    for i in range(len(hist)):
#         print(f"Intervalo {i+1}: {hist[i]} datos")
        if hist[i]>m:
            m=hist[i]
    return m
```
