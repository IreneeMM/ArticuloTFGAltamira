# Índice: Qué hay en este fichero

 ¿Las oscilaciones diarias se deben a un intercambio de aire con el exterior?
- CCs sin suavizar datos de la diferencia de temperatura exterior y la de las salas de interes con la cantidad de CO2 de la sala
- CCs tras suavizar datos de la diferencia de temperatura exterior y la de las salas de interes con la cantidad de CO2 de la sala

Significado de las abreviaturas:
- CC: Coeficientes de correlación
- CC_difT_CO2_cruce: Coeficiente de correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en el Cruce
- pvalue_difT_CO2_cruce: P-valor de la correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en el Cruce
- CC_difT_CO2_hall: Coeficiente de correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en el Hall
- pvalue_difT_CO2_hall: P-valor de la correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en el Hall
- CC_difT_CO2_policromos: Coeficiente de correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en Policromos
- pvalue_difT_CO2_policromos: P-valor de la correlación del gradiente de temperatura(T exterior - T de la sala) y la cantidad de CO2 en Policromos
- nº datos: Número de filas (pares de datos) utilizadas en cada correlación
